#### Notebook to combine all feature engineered data and merge it with all non-feature engineered data

```python
import pandas as pd
import os
pd.set_option('display.max_columns', None)
pd.options.display.float_format = '{:.15f}'.format ## see 15 decimal places of the numbers
from functools import reduce
```

```python
df_append=[]
for i in range (2003, 2020):
    year=str(i)
    
    PATH_1=os.path.join('..','all_years','CAMS', year ,'feature_engineering_90th')

    # list all the pkl files in the folder
    pkl_files = [file for file in os.listdir(PATH_1) if file.endswith('.pkl')]

    ## read all the pickle files
    dataframes = [pd.read_pickle(os.path.join(PATH_1, file)) for file in pkl_files]

    ## remove the NAME columns from the dataframes. This step was necessary because, functools (below) didnt work
    ## when dataframes had same columns

    for j in range(len(dataframes)):
        if "NAME" in dataframes[j].columns:
            dataframes[j] = dataframes[j].drop(columns="NAME")
   
    #from functools import reduce. merge all the datframes in the list on fips
    df1=reduce(lambda x, y: pd.merge(x, y, on = 'fips'), dataframes)
    df1['year']=int(i) ## create a column with year as value


    PATH_2=os.path.join('..','all_years','CAMS', year ,'feature_engineering','PM10_above_threshold.pkl')
    df2=pd.read_pickle(PATH_2) ## read the previous data
    df2 = df2.drop('NAME', axis=1)

    df_12=pd.merge(df1, df2, on=['fips']) 


    PATH_3=os.path.join('..','all_years','CAMS', year ,'feature_engineering','temp_above_threshold.pkl')
    df3=pd.read_pickle(PATH_3) ## read the previous data
    df3 = df3.drop('NAME', axis=1)

    df12_3=pd.merge(df_12, df3, on=['fips']) 

    PATH_4=os.path.join('..','all_years','CAMS', year ,'feature_engineering','temp_below_threshold.pkl')
    df4=pd.read_pickle(PATH_4) ## read the previous data
    df4 = df4.drop('NAME', axis=1)

    df12_4=pd.merge(df12_3, df4, on=['fips']) 
    df12_4

    PATH_5=os.path.join('..','all_years','CAMS', year ,'FE_PM25_'+ year +'.pkl') ## add the pm2.5 data
    df5=pd.read_pickle(PATH_5) ## read the previous data
    df5 = df5.drop('NAME', axis=1)

    df12_5=pd.merge(df12_4, df5, on=['fips'])
    df12_5

    df_append.append(df12_5)

full_df=pd.concat(df_append,ignore_index=True)
```
