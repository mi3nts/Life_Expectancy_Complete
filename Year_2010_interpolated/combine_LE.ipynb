#### Notebook to combine the life expectency in a county with the variables of a county.

```python
## conda environment Weather_Prediction

import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import geopandas as gpd
import xarray as xr
```

```python
## load the data with county and corresponding variables

county_var=pd.read_pickle('county_variables.pkl')
county_var.head()
```

```python
## The life expectency data

LE_DATA=pd.read_csv('LE_2010.csv')
LE_DATA
```

```python
## merge life expectency data with the variables by fips

final_df=LE_DATA.merge(county_var,how='left', on='fips')
final_df=final_df.dropna()
final_df
```
