```python
## conda environment Weather_Prediction

import os
import numpy as np
import pandas as pd
import geopandas as gpd
import xarray as xr
pd.set_option('display.max_columns', None)
```

```python
## get single level variables

PATH_1=os.path.join('../Data','CAMS','2010','2010_single_level_34_variables.nc')
PATH_2=os.path.join('../Data','CAMS','2010','2010_single_level_2_variables.nc')
sl_34_variables = xr.open_dataset(PATH_1) ## single level 34 variables
sl_2_variables = xr.open_dataset(PATH_2) ## single level remaning 2 variables
sl_34_variables
```

```python
sl_2_variables
```

```python
## resample the data to 1 year average

year_avg_sl_34 = sl_34_variables.resample(valid_time='1YE').mean(dim='valid_time')
year_avg_sl_2 = sl_2_variables.resample(valid_time='1YE').mean(dim='valid_time')
year_avg_sl_34
```

```python
year_avg_sl_34.to_dataframe() ## note that even though the valid time says, 2010-12-31, its the 1 year average
```

```python
year_avg_sl_2.to_dataframe()
```

```python
single_level=xr.merge([year_avg_sl_34,year_avg_sl_2])
single_level
```

```python
single_level.to_dataframe()
```

```python
## get multi-level variables

PATH_3=os.path.join('../Data','CAMS','2010','2010_multi_level_26_variables.nc')
ml_26_variables = xr.open_dataset(PATH_3) ## multi level 26 variables
ml_26_variables
```

```python
## resample data to 1 year

year_avg_ml_26 = ml_26_variables.resample(valid_time='1YE').mean(dim='valid_time')
year_avg_ml_26 = year_avg_ml_26.squeeze(dim="model_level") ## remove the model_level dimension
year_avg_ml_26
```

```python
year_avg_ml_26.to_dataframe()
```

```python
## note that even though the correct coordinates were downloaded from the website...
## the coordinate values of the multi-level variables are different from the ones that were selected
## we can substract 360 from the longitude values using year_avg_ml_26['longitude'] = year_avg_ml_26['longitude']-360,
## and then merge it, but doing so did not work
## so we will just merge it with the dataarray

year_avg_1=single_level.merge(year_avg_ml_26,join='override')
year_avg_1
```

```python
year_avg_1=year_avg_1.drop_vars("model_level")
year_avg_1.to_dataframe()
```

```python
## get monthly averaged data

PATH_4=os.path.join('../Data','CAMS','2010','2010_monthly_averaged_19_variables.nc')
monthly_19_variables = xr.open_dataset(PATH_4) ## multi level 26 variables
monthly_19_variables
```

```python
## resample data to 1 year

year_avg_monthly_19 = monthly_19_variables.resample(valid_time='1YE').mean(dim='valid_time')
year_avg_monthly_19
```

```python
year_avg_monthly_19.to_dataframe()
```

```python
year_avg=xr.merge([year_avg_1,year_avg_monthly_19])
year_avg
```

```python
year_avg.to_dataframe()
```
